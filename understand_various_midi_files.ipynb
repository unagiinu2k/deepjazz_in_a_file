# if instruments are stored

```python
import glob
```

```python
with_instruments = list()
no_instruments = list()
for file in glob.glob("chorales/midi/*.mid"):
    midi = converter.parse(file)
    
    parts = instrument.partitionByInstrument(midi)
    #parts = instrument.partitionByIntrument(midi)
    if parts:
        with_instruments.append(file)
    else:
        no_instruments.append(file)
```

```python
print("no. of files with instruments : {}".format(len(with_instruments)))
print("no. of files without instruments : {}".format(len(no_instruments)))
```

# number of instruments in a file

```python
ns = list()
files = list()
for file in glob.glob("chorales/midi/*.mid"):
    midi = converter.parse(file)
    
    repartitioned = instrument.partitionByInstrument(midi)
    #parts = instrument.partitionByIntrument(midi)
    
    if repartitioned:
                                  
        n = len([x for x in repartitioned.parts])
    else:
        n = None
    ns.append(n)
    files.append(file)
```

```python
import numpy as np
import pandas as pd
```

```python
df_n = pd.DataFrame({'file':files, 'n':ns})
```

```python
df_n.query('not n.isnull()')
```

```python
df_n.query('n > 0') 
```
